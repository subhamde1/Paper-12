```python
import math
import matplotlib.pyplot as plt

def prime_pi(x):
    """Estimates the prime counting function Ï€(x) using an approximation."""
    if x < 2:
        return 0
    # Simple approximation of Ï€(x)
    return int(x / math.log(x))

def K(x):
    """Computes the function K(x) as defined."""
    e = math.e
    log_x = math.log(x)
    pi_x = prime_pi(x)
    pi_x_e = prime_pi(x / e)
    pi_x_e2 = prime_pi(x / e**2)
    pi_x_e3 = prime_pi(x / e**3)

    term1 = pi_x**4
    term2 = 4 * e * x / log_x * pi_x_e**3
    term3 = 6 * e**2 * x / (log_x**2) * pi_x_e2**2
    term4 = 4 * e**3 * x / (log_x**3) * pi_x_e3

    return term1 - term2 + term3 - term4

# Define the range of x values
x_values = range(20000, 100001, 1000)
k_values = [K(x) for x in x_values]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(x_values, k_values, label=r'$\mathcal{K}(x)$')
plt.xlabel('x')
plt.ylabel(r'$\mathcal{K}(x)$')
plt.title(r'Plot of $\mathcal{K}(x)$ for $20000 \leq x \leq 100001$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def prime_pi(x):
    """Estimates the prime counting function Ï€(x) using an approximation."""
    if x < 2:
        return 0
    # Simple approximation of Ï€(x)
    return int(x / math.log(x))

def L(x):
    """Computes the function L(x) as defined."""
    e = math.e
    log_x = math.log(x)
    sum_pi_x_k = sum(prime_pi(x / k) for k in range(1, 6))
    sum_pi_x_ek = sum(prime_pi(x / (e * k)) for k in range(1, 6))

    term1 = sum_pi_x_k**2
    term2 = e * x / log_x * sum_pi_x_ek

    return term1 - term2

# Define the range of x values
x_values = range(20000, 100001, 1000)
l_values = [L(x) for x in x_values]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(x_values, l_values, label=r'$\mathcal{L}(x)$')
plt.xlabel('x')
plt.ylabel(r'$\mathcal{L}(x)$')
plt.title(r'Plot of $\mathcal{L}(x)$ for $20000 \leq x \leq 100001$')
plt.legend()
plt.grid(True)
plt.show()
```
